```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np



%matplotlib inline
```

```python
file_0='MI_DiagnosticsDataLog_10_11_2016.csv'
file_1='MI_DiagnosticsDataLog_10_18_2016.csv'
file_2='MI_DiagnosticsDataLog_10_19_2016.csv'
file_3='MI_DiagnosticsDataLog_10_20_2016.csv'
file_4='MI_DiagnosticsDataLog_10_21_2016.csv'
file_list=[file_0, file_1, file_2, file_3, file_4]
data_list=[]
```

```python
for i in range(5):
    data=pd.read_csv(file_list[i])
    data=data[[' Relative Time',' Platform Temp.',' Vacuum Pressure', ' Units']]
    data.columns=['Relative-time', 'Platform-temp', 'Vacuum-pressure', 'Units']    #get rid of the space in front of the column name

    #add a new column that change the timedelta to mins (pd.to_timedelta() method)
    data['Relative-time-min']=pd.to_timedelta(data['Relative-time']) / np.timedelta64(1, 'm') 

    #rearragne the columns
    data=data[['Relative-time', 'Relative-time-min','Platform-temp', 'Vacuum-pressure', 'Units']]

    cols=data.columns.tolist()

    data['Vacuum-pressure']=pd.to_numeric(list(map(lambda x: 0.001 if x==' HiVac' else x.strip(),
                                                   data['Vacuum-pressure'].tolist())))
    data['Platform-temp']=pd.to_numeric(data['Platform-temp'])


    reunits=list(map(lambda x: x[1]*1000 if x[0]==' Torr' else x[1],
                     zip(data['Units'], data['Vacuum-pressure'])))


    #label=list(map(lambda x: 1 if x==' Torr' else 0, data[' Units']))
    #reunits=list(map(lambda x: x[1]*1000 if x[0]==1 else x[1],list(zip(label, vacuum_data['Vacuum-pressure']))))
    data['Vacuum-pressure']=reunits
    data['Units']='mTorr'
    data_list.append(data)
```
